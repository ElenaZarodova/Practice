# Задание 0

Создайте функцию по нахождению уникальных элементов из двух списков



```
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
```



````python
def find_unique_elements(list1, list2):
    unique_in_list1 = []
    for x in list1:
      if x not in list2:
        unique_in_list1.append(x)
    unique_in_list2 = []
    for x in list2:
      if x not in list1:
        unique_in_list2.append(x)
    return unique_in_list1 + unique_in_list2


a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
result = find_unique_elements(a, b)
print(result)
````

# Задание 1

Симулятор виртуального питомца

Цель: создать виртуальный симулятор домашних животных, в котором пользователи смогут заводить питомцев и ухаживать за ними.

Требования:

- Функция для усыновления питомца (имя, тип, возраст).
- Функция для того, чтобы покормить питомца, поиграть с ним или уложить его спать.
- Функция для отображения состояния питомца (голод, радость, энергия).

````python
import random
import time


def create_pet():
    name = input("Введите имя питомца: ")
    pet_type = input("Введите тип питомца: ")
    age = int(input("Введите возраст питомца: "))
    return {"name": name, "type": pet_type, "age": age,
            "hunger": 50, "happiness": 50, "energy": 50, "health": 100}


def feed(pet):
    if pet["hunger"] < 100:
        pet["hunger"] = max(pet["hunger"] - 10, 0)
        pet["health"] = min(pet["health"] + 5, 100)
        print(f"{pet['name']} покормлен."
              f"Голод: {pet['hunger']}, здоровье: {pet['health']}")
    else:
        print(f"{pet['name']} не голоден.")


def play(pet):
    if pet["energy"] >= 10:
        pet["happiness"] = min(pet["happiness"] + 10, 100)
        pet["energy"] -= 10
        print(f"{pet['name']} поиграл."
              f"Радость: {pet['happiness']}, Энергия: {pet['energy']}")
    else:
        print(f"{pet['name']} слишком устал, чтобы играть.")


def sleep(pet):
    pet["energy"] = min(pet["energy"] + 20, 100)
    print(f"{pet['name']} поспал. Энергия: {pet['energy']}")


def care(pet):
    pet["health"] = min(pet["health"] + 10, 100)
    print(f"{pet['name']} ухожен. Здоровье: {pet['health']}")


def show_status(pet):
    print(f"\nСтатус питомца {pet['name']}:")
    print(f"Тип: {pet['type']}, Возраст: {pet['age']}")
    print(f"Голод: {pet['hunger']}, Радость: {pet['happiness']},"
          f"Энергия: {pet['energy']}, Здоровье: {pet['health']}")
    if pet["hunger"] <= 10:
        print(f"{pet['name']} очень голоден!")
    if pet["happiness"] <= 10:
        print(f"{pet['name']} грустит!")
    if pet["energy"] <= 10:
        print(f"{pet['name']} устал!")
    if pet["health"] <= 10:
        print(f"{pet['name']}чувствует себя плохо, нужно заботиться о здоровье")


def random_event(pet):
    event = random.choice(["illness", "happy_day", "none"])
    if event == "illness":
        print(f"{pet['name']} заболел! Здоровье снизилось.")
        pet["health"] = max(pet["health"] - 20, 0)
    elif event == "happy_day":
        print(f"{pet['name']} радуется! Радость повысилась.")
        pet["happiness"] = min(pet["happiness"] + 10, 100)


def main():
    print ("Привет, это Симулятор виртуального питомца")
    my_pet = create_pet()

    while True:
        show_status(my_pet)
        random_event(my_pet)

        print ("Выбери действие")
        print ("1 - Кормление")
        print ("2 - Игра")
        print ("3 - Уложить спать питомца")
        print ("4 - Состояние питомца")
        print ("5 - Окончание программы ")
        action = int(input("Что вы хотите сделать с питомцем?: "))

        if action == 1:
            feed(my_pet)
        elif action == 2:
            play(my_pet)
        elif action == 3:
            sleep(my_pet)
        elif action == 4:
            care(my_pet)
        elif action == 5:
            print(f"{my_pet['name']} прощается с вами!")
            break
        else:
            print("Неизвестная команда. Попробуйте снова.")



if __name__ == "__main__":
    main()
````

# Задание 2

Рыцарь и дракон

Цель: создать небольшую игру, в которой вам необходимо играть за рыцаря и сразиться с драконом

Требования:

- Создание персонажа (имя, информация о доспехах, оружии, урон, здоровье)
- Управление персонажем и мини сюжет
- Создание дракона (Имя, информация о здоровье и уроне)
- Боевая система (нанесение и получение урона, урон должен быть случайным в заданном диапазоне)
- Реализовать бой между драконом и рыцарем

````python
import random


knight = {
    "name": input("Введите имя рыцаря: "),
    "armor": "Кольчужный доспех",
    "weapon": "Меч",
    "damage_range": (10, 20),
    "health": 100
}


dragon = {
    "name": "Огнекрыл",
    "damage_range": (5, 15),
    "health": 150
}


def attack(attacker, defender):
    damage = random.randint(*attacker["damage_range"])
    defender["health"] -= damage
    print(f"{attacker['name']} наносит {damage} урона. "
          f"У {defender['name']} осталось {max(defender['health'], 0)} здоровья.")


print(f"\n{knight['name']} в доспехах ({knight['armor']}) "
      f"и с оружием ({knight['weapon']}) сражается с драконом {dragon['name']}!")
while knight["health"] > 0 and dragon["health"] > 0:
    input("\nНажмите Enter для атаки...")
    attack(knight, dragon)
    if dragon["health"] <= 0:
        print(f"\n{dragon['name']} побежден! {knight['name']} одержал победу!")
        break

    attack(dragon, knight)
    if knight["health"] <= 0:
        print(f"\n{knight['name']} пал в бою. {dragon['name']} одержал победу!")
        break
````

# Задание 3


Цель - создать менеджера команды Pokémon, который позволит пользователям:

- Добавлять покемонов в свою команду. (если такого покемона еще нет в команде)
- Удалять покемонов из их команды.
- Просматривать подробную информацию обо всех покемонах в команде.
- Находить покемона по имени.
- Устраивать тренировочный бой между двумя покемонами

Для данной задачи используйте: https://pokeapi.co/


````python
import requests


def get_pokemon_data(name):
    response = requests.get(f"https://pokeapi.co/api/v2/pokemon/{name.lower()}")
    if response.status_code == 200:
        data = response.json()
        return {
            "name": data["name"],
            "hp": data["stats"][0]["base_stat"],
            "attack": data["stats"][1]["base_stat"],
            "defense": data["stats"][2]["base_stat"],
            "type": [t["type"]["name"] for t in data["types"]]
        }
    else:
        print(f"Покемон {name} не найден.")
        return None


def add_pokemon(team, name):
    if name in team:
        print(f"{name} уже есть в команде.")
    else:
        data = get_pokemon_data(name)
        if data:
            team[name] = data
            print(f"{name} добавлен в команду.")


def remove_pokemon(team, name):
    if name in team:
        del team[name]
        print(f"{name} удален из команды.")
    else:
        print(f"{name} отсутствует в команде.")


def show_team(team):
    if team:
        for name, data in team.items():
            print(f"\nИмя: {data['name']}, Здоровье: {data['hp']},"
                  f"Атака: {data['attack']}, Защита: {data['defense']},"
                  f"Тип: {', '.join(data['type'])}")
    else:
        print("Команда пуста.")


def battle(pokemon1, pokemon2):
    print(f"Бой начинается: {pokemon1['name']} против {pokemon2['name']}!")
    damage1 = random.randint(pokemon1['attack'] - 5, pokemon1['attack'] + 5)
    damage2 = random.randint(pokemon2['attack'] - 5, pokemon2['attack'] + 5)

    pokemon2['hp'] -= max(damage1 - pokemon2['defense'], 0)
    pokemon1['hp'] -= max(damage2 - pokemon1['defense'], 0)

    print(f"{pokemon1['name']} наносит {damage1} урона {pokemon2['name']}. "
          f"У {pokemon2['name']} осталось {pokemon2['hp']} здоровья.")
    print(f"{pokemon2['name']} наносит {damage2} урона {pokemon1['name']}."
          f"У {pokemon1['name']} осталось {pokemon1['hp']} здоровья.")

    if pokemon1['hp'] <= 0 and pokemon2['hp'] <= 0:
        print("Бой завершился ничьей!")
    elif pokemon1['hp'] <= 0:
        print(f"{pokemon2['name']} побеждает!")
    elif pokemon2['hp'] <= 0:
      print ("Бой продолжается")


team = {}
def main():
  while True:
      print ("Введите одну из выбранных команд")
      print ("1 - Добавить покемона")
      print ("2 - Удалить покемона")
      print ("3 - Показать команду")
      print ("4 - Найти информацию о покемоне")
      print ("5 - Провести тестовый бой")
      print ("6 - Завершить выполнение")
      command = int(input("Введите команду: "))

      if command == 1:
          name = input("Введите имя покемона для добавления: ").strip().lower()
          add_pokemon(team, name)

      elif command == 2:
          name = input("Введите имя покемона для удаления: ").strip().lower()
          remove_pokemon(team, name)

      elif command == 3:
          show_team(team)

      elif command == 4:
          name = input("Введите имя покемона для поиска: ").strip().lower()
          if name in team:
              data = team[name]
              print(f"Найден {name} — Здоровье: {data['hp']}, "
              f"Атака: {data['attack']}, Защита: {data['defense']},"
              f"Тип: {', '.join(data['type'])}")
          else:
              print(f"{name} не найден в команде.")
      elif command == 5:
          if len(team) < 2:
              print("В команде должно быть как минимум два покемона для боя.")
          else:
              pokemon_names = list(team.keys())
              pokemon1_name = pokemon_names[0]
              pokemon2_name = pokemon_names[1]
              pokemon1 = team[pokemon1_name]
              pokemon2 = team[pokemon2_name]
              battle(pokemon1, pokemon2)

      elif command == 6:
          print("Завершение программы.")
          break

      else:
          print("Неизвестная команда. Попробуйте снова.")

if __name__ == "__main__":
    main()
````
