# **Задача 1.**




Типы данных Напишите программу на Python, которая принимает список чисел в качестве входных данных от пользователя, а затем выводит сумму, среднее значение и максимальное значение из списка.

***Нельзя использовать: len(), min(), max(), average(), sum()***

Ввод:
```
1 2 3 4 5 6 7
```
Вывод:

```
Сумма чисел: 28
Среднее арифм.: 4.0
Максимальное значение: 7
```



````python
num0=input('введите числа: ')
num1=num0.split()
sm=mx=k=0
for i in num1:
  if mx<int(i):
    mx=int(i)
  k+=1
  sm+=int(i)
print('Сумма', sm)
print('Среднее значение',sm/k)
print('Максимальный', mx)

````

# **Задача 4.**


Камень, ножницы, бумага

Создайте игру в Камень, ножницы, бумагу, в которой пользователь может играть против компьютера. Программа должна запросить выбор пользователя, а затем сгенерировать выбор компьютера. Затем программа должна определить победителя на основе правил игры.

**Этап 1:**

Определение победителя

**Этап 2:**
  
Игра до 3 побед


Ввод для "Этап 1":

```
Камень
```

Вывод:

```
Компьютер: Ножницы
Игрок: Камень
Победа: Игрок
```

````python
import random


def win(player, comp):
  if comp==player:
    return 'Ничья'
  elif (player=='Ножницы' and comp=='Бумага') or (player=='Камень' and comp=='Ножницы') or (player=='Бумага' and comp=='Камень'):
    return 'Игрок'
  else:
    return 'Компьютер'


def game():
  option=['Камень','Ножницы','Бумага']
  player=input('Введите свой выбор: Камень, Ножницы, Бумага - ')
  if player not in option:
    print ('Не верный ввод данных')
    return None
  comp=random.choice(option)
  print(f'Компьютер: {comp}')
  winner=win(player, comp)
  if winner=='Ничья':
    print ('Ничья')
  else:
    print (f'Победитель: {winner}')
  return winner


def rounds():
  player_wins=0
  comp_wins=0
  while player_wins<3 and comp_wins<3:
    winner=game()
    if winner=='Игрок':
      player_wins+=1
    elif winner=='Компьютер':
      comp_wins+=1
    print (f'Счет- игрок:{player_wins}', f'компьютер:{comp_wins}')
  if player_wins==3:
    print ('Игрок выиграл')
  else:
    print ('Компьютер выиграл')
rounds()


````

# **Задача 5.**

Создайте игру "Палач" или "Виселица", в которой пользователь должен угадать слово, предлагая буквы. Программа должна выбрать случайное слово, а затем разрешить пользователю угадывать буквы. После каждого угадывания программа должна сообщать пользователю, есть ли в слове буква или нет.

**Этап 1**
Создать саму игру (игра должна иметь привлекательный интерфейс, соотвесвующий классической висилице)

**Этап 2:**
Ввести подсчет количества очков и ограничения на попытки

**Этап 3:** Добавить возможность сразу угадать слово

````python
import random

words=['studient', 'protective', 'crystal' , 'truth']
secret_word = random.choice(words)
guessed_letters=[]
wrongs=0
max_attempts=8
score=10

print('сейчас мы сыграем в игру виселица')
print('ваша задача вводить буквы и угадать рандомное слово')
print('также можете попробовать угадать слово целиком')
print(f'сейчас ваши очки составляют {score}')
print('если не угадываете букву теряете 1 балл, если слово - 2 балла')
while wrongs!=max_attempts:
  print(f"Количество ошибок: {wrongs}/{max_attempts}")
  print(f"Текущие очки: {score}")
  guess=input('введите букву или слово на англ.яз маленькими буквами ')
  if len(guess)==1:
    if guess in secret_word:
      guessed_letters.append(guess)
      print (f'буква {guess} есть в слове')
    elif guess in guessed_letters:
      print('данная буква уже проверена вами')
    else:
      guessed_letters.append(guess)
      wrongs+=1
      score-=1
      print (f'буквы {guess} нет в слове')
    if all(letter in guessed_letters for letter in secret_word):
      print (f'вы выиграли и угадали слово {secret_word}')
      print (f'ваш счет составил {score}')
      break
  elif len(guess)==len(secret_word):
    if guess==secret_word:
      print (f'вы выиграли и угадали слово {secret_word}')
      print (f'ваш счет составил {score}')
      break
    else:
      wrongs+=1
      score-=2
      print ('это неверное слово')
  if wrongs==8 or score==0:
    print ('вы проиграли, у вас закончились попытки')



````
